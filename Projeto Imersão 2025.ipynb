```python
import google.generativeai as genai
import os

os.environ['GOOGLE_API_KEY'] = "AIzaSyBKIz2kRGGj8wxZ_7-rfbb0ekn7YsAqhr4"
genai.configure(api_key=os.environ['GOOGLE_API_KEY'])

model = genai.GenerativeModel('gemini-1.5-flash')
print("Usando modelo: gemini-1.5-flash")


def coletar_info_usuario():
    info = {}
    print("Por favor, responda às seguintes perguntas:")
    info['estilo_aprendizado'] = input(
        "Qual seu estilo de aprendizado predominante (visual, auditivo, leitura/escrita, cinestésico)? "
    ).lower()
    info['horarios_disponiveis'] = input(
        "Em quais horários você geralmente tem tempo para estudar (ex: seg-sex 18-20, sab 10-12)? "
    ).lower()
    info['materias_prioritarias'] = [
        m.strip()
        for m in input("Quais matérias você precisa priorizar (separe por vírgula)? ").split(",")
    ]
    info['tempo_dedicado'] = input(
        "Quanto tempo você gostaria de dedicar aos estudos por dia/semana? "
    ).lower()
    info['objetivos'] = input(
        "Quais são seus objetivos de estudo (ex: melhorar nota em matemática, aprender um novo tópico)? "
    ).lower()
    return info


def gerar_prompt_estudos(info):
    prompt = f"""
    Com base nas seguintes informações sobre o estudante:
    - Estilo de aprendizado: {info['estilo_aprendizado']}
    - Horários disponíveis: {info['horarios_disponiveis']}
    - Matérias prioritárias: {', '.join(info['materias_prioritarias'])}
    - Tempo dedicado aos estudos: {info['tempo_dedicado']}
    - Objetivos de estudo: {info['objetivos']}

    Gere um plano de estudos semanal detalhado em formato Markdown. Inclua:

    ## Plano de Estudos Semanal

    Para cada dia da semana, liste os horários de estudo e a matéria correspondente.

    ### Recursos de Aprendizado Sugeridos

    Para cada matéria prioritária ({', '.join(info['materias_prioritarias'])}), liste 2-3 recursos online relevantes para o estilo de aprendizado do estudante ({info['estilo_aprendizado']}), com um breve resumo de cada um. Use listas não ordenadas.

    ### Dicas de Técnicas de Estudo

    Forneça 2-3 dicas de técnicas de estudo que se adequem ao estilo de aprendizado do estudante ({info['estilo_aprendizado']}). Use listas não ordenadas.

    O plano de estudos deve ser realista e adaptável.
    """
    return prompt


def exibir_plano_de_estudos(plano):
    print("\nSeu Plano de Estudos Personalizado:\n")
    print(plano)


def iniciar_chat():
    print("Olá! Bem-vindo ao seu Guia de Estudos Personalizado!")
    info_usuario = coletar_info_usuario()
    prompt_gerado = gerar_prompt_estudos(info_usuario)
    try:
        response = model.generate_content(prompt_gerado)
        exibir_plano_de_estudos(response.text)
    except Exception as e:
        print(f"Ocorreu um erro ao gerar o conteúdo: {e}")


if __name__ == "__main__":
    iniciar_chat()
```
